# Import libraries

```python
import os
from pathlib import Path
import shutil
```

```python
# Define paths
DEST_DIR = "../data/converted/Tsync-to-vctk"
DEST_TEXT_PATH = os.path.join(DEST_DIR, "txt")
DEST_AUDIO_PATH = os.path.join(DEST_DIR, "wav32")
SRC_AUDIO_PATH = "../data/raw/Tsync2/wav"
SRC_TEXT_PATH = "../data/raw/Tsync2/wrd_ph"

# Clean and create directories
if os.path.exists(DEST_DIR):
   print("Clearing destination folder")
   shutil.rmtree(DEST_DIR)
os.makedirs(DEST_TEXT_PATH, exist_ok=True)
os.makedirs(DEST_AUDIO_PATH, exist_ok=True)


# Convert audio files from Tsync to VCTK format
for file in Path(SRC_AUDIO_PATH).glob("*.wav"):  # Only process .wav files
   #  data, samplerate = sf.read(file)
   #  print(f"{file.name}: {data.shape}, Sample Rate: {samplerate}")
   src_file = os.path.join(SRC_AUDIO_PATH, file.name)
   dest_file = os.path.join(DEST_AUDIO_PATH, file.name)
   shutil.copy2(src_file, dest_file)

#Convert txt files from Tsync to VXTK format
for file in Path(SRC_TEXT_PATH).glob("*.txt"):  
    with file.open('r', encoding='utf-8') as f:
        first_line = f.readline().strip()
        clean_text = "".join(first_line.split("|"))
        print(clean_text)
        
        file_path = os.path.join(DEST_TEXT_PATH, file.name)  # Define the file name
        with open(file_path, 'w', encoding='utf-8') as file:
             file.write(clean_text)
```
